17.1 Interpolation

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,5,1,6]
plt.figure(figsize = (10,8))
plt.plot(x,y,'bo', 2.5,3,'ro')
plt.title('Interpolation data point')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.2 Linear Interpolation

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt


x = [0, 1, 3, 5]
y = [1, 5, 3, 2.5]

f = interp1d(x, y)
y_hat = f(2)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(2, y_hat, 'ro')
plt.title('Linear Interpolation at x = 2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.3 Cubic Interpolation

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt


x = [0, 2, 4, 6]
y = [1, 7, 6, 4]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 7, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
